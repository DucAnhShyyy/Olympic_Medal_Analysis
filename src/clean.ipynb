#### Ở trong dataset có 2 cột gender khác nhau, đó là Gender và Event_gender. Trong đó, gender là chỉ giới tính của vận động viên giành huy chương trong môn đó, còn event_gender để chỉ giới tính được tham gia vào bộ môn đó. Nếu event_gender = M, tức là môn đó chỉ dành cho nam, tương tự W là cho nữ, còn X thì sẽ là môn mà cả 2 giới tính đều được tham gia.  

#### Ví dụ ở đoạn code phía trên để lấy ra danh sách hạng mục mà chỉ có nam được tham gia, có môn Wrestling Gre-R, tức là môn Greco-Roman Wrestling (Vật tay kiểu Greco-Roman) thì theo như tra mạng, đây là môn chỉ dành cho nam.

#### Vậy nên ko nhất thiết phải bỏ cột đấy đi, mà có thể giữ lại nếu muốn phân tích về các môn mà nam vs nữ cùng tham gia

```python
olympic_final = olympic.merge(noc, on= 'NOC', how= 'left')
new_order = ['City', 'Edition', 'Sport', 'Discipline', 'Athlete', 'NOC', 'Athlete Country', 'Gender', 'Event', 'Event_gender', 'Medal']
olympic_final = olympic_final[new_order]
olympic_final.head()
```

```python
olympic_final[olympic_final['Athlete Country'].isnull()].NOC.unique()
```

```python
# TRI là mã cũ của TTO, là nước Trinidad and Tobago
# LIB là mã cũ của LBN, tức là Lebanon
# SIN là mã cũ của SGP, tức là Singapore
# ZZX là mã cũ của XXB, tức là mixed team của các nước khác nhau, XXB được dùng từ 2024
```

```python
olympic_final.loc[olympic_final['NOC'] == 'TRI', 'Athlete Country'] = olympic_final.loc[olympic_final['NOC'] == 'TRI', 'Athlete Country'].fillna('Trinidad and Tobago')
olympic_final.loc[olympic_final['NOC'] == 'LIB', 'Athlete Country'] = olympic_final.loc[olympic_final['NOC'] == 'LIB', 'Athlete Country'].fillna('Lebanon')
olympic_final.loc[olympic_final['NOC'] == 'SIN', 'Athlete Country'] = olympic_final.loc[olympic_final['NOC'] == 'SIN', 'Athlete Country'].fillna('Singapore')
olympic_final.loc[olympic_final['NOC'] == 'ZZX', 'Athlete Country'] = olympic_final.loc[olympic_final['NOC'] == 'ZZX', 'Athlete Country'].fillna('Mixed Team')
```

```python
# Germany bao gồm Germany, West/East Germany, EUA (United Team of Germany)
# Russia bao gồm Russia, Soviet Union, Unified Team (EUN), RU1 (Russian Empire)
# Netherlands gồm Netherlands, Netherlands Antilles (AHO)
# Czechia (séc) gồm cả Czechia và Bohemia (trong lịch sử bohemia là phần của áo hung, sau này thuộc séc) 
# Yugoslavia (YUG) là lãnh thổ của Serbia và Montenegro, sau 2003 đổi tên thành Serbia and Montenegro
# BWI là tây ấn thuộc anh, gồm nhiều nước
# Australasia gồm các nc châu đại dương, coi như ko tách nữa
```

```python
olympic_final.loc[olympic_final['NOC'].isin(['FRG', 'EUA', 'GDR']), 'Athlete Country'] = 'Germany'
olympic_final.loc[olympic_final['NOC'].isin(['URS', 'RU1', 'EUN']), 'Athlete Country'] = 'Russia'
olympic_final.loc[olympic_final['NOC'].isin(['AHO']), 'Athlete Country'] = 'Netherlands'
olympic_final.loc[olympic_final['NOC'].isin(['BOH']), 'Athlete Country'] = 'Czechia'
olympic_final.loc[olympic_final['NOC'].isin(['YUG']), 'Athlete Country'] = 'Serbia and Montenegro'
```

```python
city_to_continent = {
    'Athens': 'Europe',
    'Paris': 'Europe',
    'St. Louis': 'North America',
    'London': 'Europe',
    'Stockholm': 'Europe',
    'Antwerp': 'Europe',
    'Amsterdam': 'Europe',
    'Los Angeles': 'North America',
    'Berlin': 'Europe',
    'Helsinki': 'Europe',
    'Melbourne': 'Oceania',
    'Rome': 'Europe',
    'Tokyo': 'Asia',
    'Mexico City': 'North America',
    'Munich': 'Europe',
    'Montreal': 'North America',
    'Moscow': 'Europe',
    'Seoul': 'Asia',
    'Barcelona': 'Europe',
    'Atlanta': 'North America',
    'Sydney': 'Oceania',
    'Beijing': 'Asia'
}

country_host['Continent'] = country_host['City'].map(city_to_continent)
country_host
```

```python
country_replacements = {
    "West Germany (now Germany)": "Germany",
    "U.S.S.R. (now Russia)": "Russia"}

country_host['Country'] = country_host['Country'].replace(country_replacements)
country_host.rename(columns= {'Country': 'Host Country'}, inplace= True)
```

```python
olympic_final = olympic_final.drop(columns= ['City', 'NOC']).merge(country_host, on= ['Edition'], how= 'left') 
# => Cột city có vài thành phố viết sai, nên chỉ lấy cột city ở bảng country host
```
